## Estatísticas de Carreira: Desempenho de LeBron James

```python
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.static import players

# Defina o nome do jogador
nome_jogador = "LeBron James"

# Encontrar o ID do jogador usando o nome
jogadores = players.get_players()
jogador = next(player for player in jogadores if player['full_name'] == nome_jogador)
id_jogador = jogador['id']

# Obter as estatísticas de carreira do jogador
carreira = playercareerstats.PlayerCareerStats(player_id=id_jogador)

# Acessar os dados em formato DataFrame
df_carreira = carreira.get_data_frames()[0]
df_carreira
```

## Registros de Jogo: Detalhes das Partidas da Temporada

```python
from nba_api.stats.endpoints import leaguegamelog

gamelog = leaguegamelog.LeagueGameLog(season="2023-24", season_type_all_star='Regular Season')

# Acessar os dados em formato DataFrame
df_gamelog = gamelog.get_data_frames()[0]
df_gamelog
```

## Desempenho das Equipes: Análise das Estatísticas dos Times

```python
from nba_api.stats.endpoints import leaguedashteamstats

# Obter as estatísticas das equipes para a temporada 2023-24
team_stats = leaguedashteamstats.LeagueDashTeamStats(season="2023-24")

# Acessar os dados em formato DataFrame
df_team_stats = team_stats.get_data_frames()[0]
df_team_stats
```
